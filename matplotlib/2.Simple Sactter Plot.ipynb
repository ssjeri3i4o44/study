plt.scatter与plt.plot的主要区别在于，它可用于创建散点图，其中每个单独点的属性（大小、面颜色、边颜色等）可单独控制或映射到数据。

让我们通过创建一个具有许多颜色和大小的点的随机散点图来说明这一点。为了更好地看到重叠的结果，我们还将使用alpha关键字来调整透明度级别：

```python
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000*rng.rand(100)

plt.scatter(x, y, c = colors, s = sizes, alpha = 0.5, cmap = 'viridis')
plt.colorbar()#展示颜色,将color参数自动映射到一个色阶
```

用鸢尾花的例子来展示三种花的花瓣和萼片大小

```python
from sklearn.datasets import load_iris
iris = load_iris()
features = iris.data.T
plt.scatter(features[0], features[1], alpha = 0.5,
            s = 100*features[3],c = iris.target,cmap = 'viridis')
# s:size,表示点的大小；c:color，表示点的颜色
# cmap 色彩映射
print(iris.target)
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
```
