```python
import numpy as np
```

```python
#使用numpy构建一个单隐层神经网络
N, D_in, H, D_out = 64, 100, 100, 10
#N = batchsize, H = hide layer, D_in = input, D_out = output
#define x and y
x = np.random.randn(N, D_in)
y = np.random.randn(N, D_out)

#define w1 and w2
w1 = np.random.randn(D_in, H)
w2 = np.random.randn(H, D_out)

lr = 0.000001

for i in range(400):
    #define forward
    h = np.dot(x, w1)
    h_relu = np.maximum(h, 0)
    y_pre = np.dot(h_relu, w2)
    
    #loss function
    loss = np.square(y_pre - y).sum()
    print(i, loss)
    if loss < 1:
        break
    #define backward
    grad_y = 2.0*(y_pre - y)
    grad_w2 = np.dot(h_relu.T, grad_y)
    grad_h_relu = np.dot(grad_y, w2.T)
    grad_h = grad_h_relu.copy()
    grad_h[h > 0] = 0
    grad_w1 = np.dot(x.T, grad_h)
    
    #updata weights
    w1 -= lr*grad_w1
    w2 -= lr*grad_w2
```

```python
#use pytorch to fit two layer nerual network
import torch
```

```python
dtype = torch.float
device = torch.device('cpu')
#device = torch.device('cuda:0')#Uncomment this to run on gpu

N, D_in, H, D_out = 64, 100, 100, 10

x = torch.randn(N, D_in)
y = torch.randn(N, D_out)

model = torch.nn.Sequential(
        torch.nn.Linear(D_in, H),
        torch.nn.ReLU(),
        torch.nn.Linear(H, D_out))

loss_fn = torch.nn.MSELoss(reduction='sum')

lr = 0.0001

optimizer = torch.optim.Adam(model.parameters(), lr = lr)

for i in range(500):
    y_pred = model(x)
    loss = loss_fn(y_pred,y)
    print(i, loss.item())
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
```

```python
dtype = torch.float
device = torch.device('cpu')

N,D_in, H, D_out = 64, 100, 100, 10

x = torch.randn(N, D_in, device=device, dtype=dtype)
y = torch.randn(N, D_out, device=device, dtype=dtype)

w1 = torch.randn(D_in, H, device=device, dtype=dtype, requires_grad=True)
w2 = torch.randn(H, D_out, device=device, dtype=dtype, requires_grad=True)

lr = 0.00001
for t in range(500):
    y_pred = x.mm(w1).clamp(min=0).mm(w2)
    
    loss = (y_pred * y).pow(2).sum()
    print(t, loss.item())
    
    loss.backward()
    
    with torch.no_grad():
        w1 -= lr * w1.grad
        w2 -= lr * w2.grad
        
        w1.grad.zero_()
        w2.grad.zero_()
```

```python
#define a two-layer network
import torch

class twolayernet(torch.nn.Module):
    def __init__(self, D_in, H, D_out):
        super(twolayernet, self).__init__()
        self.linear1 = torch.nn.Linear(D_in, H)
        self.linear2 = torch.nn.Linear(H, D_out)
        
    def forward(self, x):
        h_relu = self.linear1(x).clamp(min=0)
        y_pred = self.linear2(h_relu)
        return y_pred
    
    
dtype = torch.float
device = torch.device('cpu')
N, D_in, H, D_out = 64, 100, 100, 10

x = torch.randn(N, D_in, device=device, dtype=dtype)
y = torch.randn(N, D_out, device=device, dtype=dtype)

model = twolayernet(D_in, H, D_out)

loss_fn = torch.nn.MSELoss(reduction='sum')

optimizer = torch.optim.SGD(model.parameters(), lr=0.00001, momentum=0.99)

for i in range(500):
    y_pred = model(x)
    
    loss = loss_fn(y_pred, y)
    print(i, loss.item())
    
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
```
