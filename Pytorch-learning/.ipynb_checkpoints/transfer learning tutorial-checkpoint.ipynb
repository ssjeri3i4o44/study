```python
data_transforms = {
    'train':transforms.Compose([#复合图像预处理
        transforms.RandomResizedCrop(224),#先随机裁剪，然后resize
        transforms.RandomHorizontalFlip(),#按概率翻转，默认0.5
        transforms.ToTensor(),#将[H,W,C]转化成[C,H,W]的tensor
        transforms.Normalize([0.485,0.456,0.406],[0.229,0.224,0.225])
    ]),#对RGB进行归一化，给定均值和方差
    'val':transforms.Compose([
        transforms.RandomResizedCrop(224),
        transforms.RandomHorizontalFlip(),
        transforms.ToTensor(),
        transforms.Normalize([0.485,0.456,0.406],[0.229,0.224,0.225])
    ])
}

data_dir= 'E:\\code\\pytorch\\hymenoptera_data\\'
image_datasets = {x:datasets.ImageFolder(os.path.join(data_dir, x),#os.path.join进行路径的拼接
                                        data_transforms[x])
                 for x in ['train', 'val']}#使用字典保存数据，写法很简洁
dataloaders = {x:torch.utils.data.DataLoader(image_datasets[x], batch_size=4,
                                             shuffle=True)
              for x in ['train', 'val']}
#将数据按批次加载，shuffle为True是指随机抽取又放回，还有一个num_workers参数，是设定多进程，同样，土豪随意
dataset_sizes = {x:len(image_datasets[x]) for x in ['train', 'val']}
print(dataset_sizes)
class_names = image_datasets['train'].classes#数据集中每个文件名代表一个类别
print(class_names)
device = torch.device('cuda:0' if torch.cuda.is_available() else 'cpu')
#我使用cpu跑的可以看到结果速度比较慢，大佬自便
```

```python
def imshow(inp, title=None):
    #'imshow for tensors'
    inp = inp.numpy().transpose((1,2,0))
    mean = np.array([0.485, 0.456, 0.406])
    std = np.array([0.229,0.224,0.225])
    inp = std*inp+mean
    inp = np.clip(inp,0,1)#设置边界，将Inp设置在[0,1]
    plt.imshow(inp)
    if title is not None:
        plt.title(title)
    plt.pause(0.001)#pause a bit so that plots are updated
    
#get a batch of training data
inputs, classes = next(iter(dataloaders['train']))
#pytorch中的torch.utils.data.Dataoader是一个组合数据集和采样器，并在数据集上提供单进程和多进程迭代器
out = torchvision.utils.make_grid(inputs)

imshow(out, title=[class_names[x] for x in classes])
```

从图片里可以看出，因为图像的预处理随机裁剪，对样本的检测物有很多缺失裁剪，目前有一些改进的裁剪方法：
1：简单的将裁剪尺寸固定到一个较大的范围，相当于稍微减少了裁剪的随机性
2：海康威视研究院提出一种有监督的裁剪，使用网络提出目标物的区域概率，然后映射到原图上进行一个裁剪
3：在SSD中，通过将原图缩小放到一个画布上再去裁剪达到缩小的效果，具体的操作可以查看论文，这种裁剪方式提高了SSD的小目标检测概率

```python
def train_model(model, criterion, optimizer, scheduler, num_epochs=25):
    since = time.time()
    best_model_wts = copy.deepcopy(model.state_dict())#可以看一下copy和deep.copy的区别
    best_acc = 0.0
    
    for epoch in range(1,num_epochs):
        print('Epoch {}/{}'.format(epoch,num_epochs))
        print('_'*10)
        
        #epoch has a training and validation phase
        for phase in ['train', 'val']:
            if phase == 'train':
                scheduler.step()#学习率根据step进行衰减
                model.train()
            else:
                model.eval()
                
            running_loss = 0.0
            running_corrects = 0.0
            
            #iterate over data
            for inputs, labels in dataloaders[phase]:
                inputs = inputs.to(device)
                labels = labels.to(device)
                
                #zero the parameter gradients
                optimizer.zero_grad()
                
                #forward
                #track history if only in train
                with torch.set_grad_enabled(phase == 'train'):
                    #根据返回的布尔值判断是否需要进行训练
                    outputs = model(inputs)
                    _, preds = torch.max(outputs.data, 1)
                    #这个就是将所有可能的分类置信度写出来，选出最大的，参数1是dim的意思
                    loss = criterion(outputs, labels)
                    
                    #backward + optimize only if in training phase
                    if phase == 'train':
                        loss.backward()
                        optimizer.step()
                #statistics
                running_loss += loss.item()*inputs.size(0)
                running_corrects += torch.sum(preds == labels.data)
                
            epoch_loss = running_loss/dataset_sizes[phase]
            epoch_acc = running_corrects.double() / dataset_sizes[phase]
            
            print('{} loss: {:.4f} Acc: {:.4f}'.format(
            phase, epoch_loss, epoch_acc))
            
            #deep copy the model
            if phase == 'val' and epoch_acc > best_acc:
                best_acc = epoch_acc
                best_model_wts = copy.deepcopy(model.state_dict())
                #保存最好精度时的参数
                
        print()
        
    over = time.time() - since
    print('training complete in {:.0f}m {:.0f}s'.format(
            over//60, over%60))
    print('best val acc:{:.4f}'.format(best_acc))
        
    #load best model weights
    model.load_state_dict(best_model_wts)
    return model
```

pytorch中保存模型的方法有两种：
1：只保存模型参数：
torch.save(the_model.state_dict(), PATH)
加载模型参数：
the_model = TheModelClass(*args, **kwargs)
the_model.load_state_dict(torch.load(PATH))

2：保存整个模型：
torch.save(the_model, PATH)
加载模型：
the_model = torch.load(PATH)

```python
def visualize_mdoel(model, num_images=6):
    was_training = model.training
    model.eval()
    image_so_far = 0
    fig = plt.figure()
    
    with torch.no_grad():
        for i, (inputs, labels) in enumerate(dataloaders['val']):
            inputs = inputs.to(device)
            labels = labels.to(device)
            outputs = model(inputs)
            _, preds = torch.max(outputs, 1)
            
            for j in range(inputs.size()[0]):
                image_so_far += 1
                ax = plt.subplot(num_images//2, 2, image_so_far)
                ax.axis('off')
                ax.set_title('predicted:{}'.format(class_names[preds[j]]))
                imshow(inputs.cpu().data[j])
                
                if image_so_far == num_images:
                    model.train(mode=was_training)
                    return
        model.train(mode=was_training)
```

```python
model_ft = models.resnet18(pretrained=True)
num_ftrs = model_ft.fc.in_features
model_ft.fc = torch.nn.Linear(num_ftrs, 2)
#将最后一层的节点
model_ft = model_ft.to(device)

criterion = torch.nn.CrossEntropyLoss()

#observe that all parameters are being optimized
optimizer_ft = torch.optim.SGD(model_ft.parameters(), lr=0.001, momentum=0.9)

#decay lr by a factor of o.1 every 7 epochs
exp_lr_scheduler = torch.optim.lr_scheduler.StepLR(optimizer_ft, step_size=2, gamma=0.1)
```

```python
model_ft = train_model(model_ft, criterion, optimizer_ft, exp_lr_scheduler, num_epochs=6)
```
