```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import datasets
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.decomposition import PCA
```

```python
iris = datasets.load_iris()
# LDA
x = iris.data
y = iris.target
target_names = iris.target_names
lda = LinearDiscriminantAnalysis(n_components=2)
x_iris = lda.fit(x,y).transform(x)
# PCA
pca = PCA(n_components=2)
X_iris = pca.fit(x,y).transform(x)
```

```python
plt.figure()
for c, i, target_name in zip("rgb", [0, 1, 2], target_names):
    plt.scatter(x_iris[y == i, 0], x_iris[y == i, 1], c=c, label=target_name)
plt.legend()
plt.title('LDA of IRIS dataset')
plt.show()
plt.figure()
for c, i, target_name in zip("rgb", [0, 1, 2], target_names):
    plt.scatter(X_iris[y == i, 0], X_iris[y == i, 1], c=c, label=target_name)
plt.legend()
plt.title('PCA of IRIS dataset')
plt.show()
```

#### PCA&LDA
PCA是一种常用的降维算法，LDA也是一种常用的降维算法：
- PCA是一种无监督降维算法，降维的根据将数据投影到最大方差方向上，也就是尽可能的保留数据的有效信息
- LDA是一种有监督降维算法，不仅能将数据降维，而且让降维后的数据更容易分类，是的类别之间距离最大，类内之间距离最小
