## 一.绘图样式(Style)
在matplotlib中，要想设置绘制样式，最简单的方法是在绘制元素时单独设置样式。
但是有时候，当用户在做专题报告时，往往会希望保持整体风格的统一而不用对每张图一张张修改，因此matplotlib库还提供了四种批量修改全局样式的方式

### 1.预先定义样式

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.style.use('Solarize_Light2')
plt.plot([1,2,3,4],[2,3,4,5])
```

```python
plt.style.use('seaborn-notebook')
plt.plot([1,2,3,4],[2,3,4,5])
```

**matplotlib中内置样式**：

```python
print(plt.style.available)
```

### 2.用户自定义样式

在任意路径下创建一个后缀名为mplstyle的样式清单，编辑文件添加以下样式内容  

> axes.titlesize : 24     
axes.labelsize : 20   
lines.linewidth : 3   
lines.markersize : 10   
xtick.labelsize : 16   
ytick.labelsize : 16  

引用自定义stylesheet后观察图表变化。

```python
plt.style.use('style/presentation.mplstyle')
plt.plot([1,2,3,4],[2,3,4,5])
```

```python
#注意:matplotlib支持混合样式的引用，只需在引用时输入一个样式列表，若是几个样式中涉及到同一个参数，右边的样式表会覆盖左边的值。
plt.style.use(['dark_background', 'style/presentation.mplstyle'])
plt.plot([1,2,3,4],[2,3,4,5])
```

### 3.设置rcparams
我们还可以通过修改默认rc设置的方式改变样式，所有rc设置都保存在一个叫做 matplotlib.rcParams的变量中。
修改过后再绘图，可以看到绘图样式发生了变化。

```python
plt.style.use('default') # 恢复到默认样式
plt.plot([1,2,3,4],[2,3,4,5])
```

```python
mpl.rcParams['lines.linewidth'] = 6
mpl.rcParams['lines.linestyle'] = '--'
plt.plot([1,2,3,4],[2,3,4,5])
```

```python
mpl.rc('lines', linewidth=1, linestyle='-.')
plt.plot([1,2,3,4],[2,3,4,5])
```

### 4.修改matplotlibrc文件
由于matplotlib是使用matplotlibrc文件来控制样式的，也就是上一节提到的rc setting，所以我们还可以通过修改matplotlibrc文件的方式改变样式。

```python
# 查找matplotlibrc文件的路径
mpl.matplotlib_fname()
```

找到路径后，就可以直接编辑样式文件了，打开后看到的文件格式大致是这样的，文件中列举了所有的样式参数，找到想要修改的参数，比如lines.linewidth: 8，并将前面的注释符号去掉，此时再绘图发现样式以及生效了。

## 二. 色彩的设置

在可视化中，如何选择合适的颜色和搭配组合也是需要仔细考虑的，色彩选择要能够反映出可视化图像的主旨。  
从可视化编码的角度对颜色进行分析，可以将颜色分为`色相、亮度和饱和度`三个视觉通道。通常来说：  
`色相`： 没有明显的顺序性、一般不用来表达数据量的高低，而是用来表达数据列的类别。  
`明度和饱和度`： 在视觉上很容易区分出优先级的高低、被用作表达顺序或者表达数据量视觉通道。  
具体关于色彩理论部分的知识，不属于本教程的重点，请参阅有关拓展材料学习。  
[ECharts数据可视化实验室](https://vis.baidu.com/chartcolor/basis/)

在matplotlib中，设置颜色有以下几种方式：

### 1.RGB或RGBA

```python
plt.style.use('default')
```

```python
# 颜色用[0,1]之间的浮点数表示，四个分量按顺序分别为(red, green, blue, alpha)，其中alpha透明度可省略
plt.plot([1,2,3],[4,5,6],color=(0.2, 0.7, 0.5))
plt.plot([4,5,6],[1,2,3],color=(0.2, 0.7, 0.5, 0.6))
```

```python
# 当只有一个位于[0,1]的值时，表示灰度色阶
plt.plot([1,2,3],[4,5,6],color='0.5')
```

### 4.单字符基本颜色

### 6.使用colormap设置一组颜色

有些图表支持使用colormap的方式配置一组颜色，从而在可视化中通过色彩的变化表达更多信息。

在matplotlib中，colormap共有五种类型:

**顺序（Sequential)**: 连续化色图，通常使用单一色调，逐渐改变亮度和颜色渐渐增加，用于表示有顺序的信息。
+ 特点：在两种色调之间近似平滑变化，通常是从低饱和度（例如白色）到高饱和度（例如明亮的蓝色）。
+ 应用：适用于大多数科学数据，可直观地看出数据从低到高的变化
>cmaps['Perceptually Uniform Sequential'] = [
            'viridis', 'plasma', 'inferno', 'magma', 'cividis']
cmaps['Sequential'] = [
            'Greys', 'Purples', 'Blues', 'Greens', 'Oranges', 'Reds',
            'YlOrBr', 'YlOrRd', 'OrRd', 'PuRd', 'RdPu', 'BuPu',
            'GnBu', 'PuBu', 'YlGnBu', 'PuBuGn', 'BuGn', 'YlGn']|
cmaps['Sequential (2)'] = [
            'binary', 'gist_yarg', 'gist_gray', 'gray', 'bone', 'pink',
            'spring', 'summer', 'autumn', 'winter', 'cool', 'Wistia',
            'hot', 'afmhot', 'gist_heat', 'copper']

```python
def plot_colormap(cmap_name):
    fig, ax = plt.subplots(figsize=(6, 2))
    cmap = mpl.cm.get_cmap(cmap_name)
    colors = cmap(np.linspace(0, 1, cmap.N))
    ax.imshow([colors], extent=[0, 10, 0, 1])
    ax.set_xticks([])
    ax.set_yticks([])
    ax.set_title(cmap_name)
```

```python
plot_colormap("viridis");
plot_colormap("plasma");
plot_colormap("inferno");
plot_colormap("magma");
plot_colormap("cividis");
plot_colormap("Greys");
plot_colormap("Purples");
plot_colormap("Greens");
plot_colormap("Oranges");
plot_colormap("Reds");
plot_colormap("YlOrBr");
plot_colormap("OrRd");
plot_colormap("PuRd");
```

- 发散（Diverging）。改变两种不同颜色的亮度和饱和度，这些颜色在中间以不饱和的颜色相遇；当绘制的信息具有关键中间值（例如地形）或数据偏离零时，应使用此值。
>cmaps['Diverging'] = [
            'PiYG', 'PRGn', 'BrBG', 'PuOr', 'RdGy', 'RdBu',
            'RdYlBu', 'RdYlGn', 'Spectral', 'coolwarm', 'bwr', 'seismic']

```python
x = np.random.randn(50)
y = np.random.randn(50)
plt.scatter(x,y,c=x,cmap='PiYG')
```

- 循环（Cyclic）。改变两种不同颜色的亮度，在中间和开始/结束时以不饱和的颜色相遇。用于在端点处环绕的值，例如相角，风向或一天中的时间。
>cmaps['Cyclic'] = ['twilight', 'twilight_shifted', 'hsv']

```python
import pandas as pd
df = pd.DataFrame([5,4,1,2,3,4,5,3,2,4], index=['a', 'b', 'c', 'd','e','f','g','h','i','j'], columns=['x'])
df.plot(kind='pie', subplots=True, figsize=(8, 8),cmap='twilight')
```

- 定性（Qualitative）。常是杂色，用来表示没有排序或关系的信息。
>cmaps['Qualitative'] = ['Pastel1', 'Pastel2', 'Paired', 'Accent',
                        'Dark2', 'Set1', 'Set2', 'Set3',
                        'tab10', 'tab20', 'tab20b', 'tab20c']

```python
x = np.random.randn(50)
y = np.random.randn(50)
plt.scatter(x,y,c=x,cmap='Pastel1')
```

```python
x = np.random.randn(50)
y = np.random.randn(50)
plt.scatter(x,y,c=x,cmap='Paired')
```

- 杂色（Miscellaneous）。一些在特定场景使用的杂色组合，如彩虹，海洋，地形等。
>cmaps['Miscellaneous'] = [
            'flag', 'prism', 'ocean', 'gist_earth', 'terrain', 'gist_stern',
            'gnuplot', 'gnuplot2', 'CMRmap', 'cubehelix', 'brg',
            'gist_rainbow', 'rainbow', 'jet', 'nipy_spectral', 'gist_ncar']

```python
df.plot(kind='pie', subplots=True, figsize=(8, 8),cmap='gist_rainbow')
```
