```python
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
```

```python
fig = plt.figure()
ax = plt.axes()
```

在matplotlib中，figure（plt.figure类的一个实例）可以看作是一个单独的容器，其中包含表示轴、图形、文本和标签的所有对象。axes（类plt.axes的一个实例）就是我们上面看到的：一个带有刻度和标签的边界框，它最终将包含构成我们可视化的绘图元素。在本书中，我们通常使用变量名fig来引用一个图实例，ax来引用一个轴实例或一组轴实例。

创建axes后，可以使用ax.plot函数绘制一些数据

```python
fig = plt.figure()
ax = plt.axes()
x = np.linspace(0, 10, 100)
ax.plot(x, np.sin(x));
```

```python
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
```

```python
import seaborn as sns
sns.plot(x, np.sin(x))
```

### 调整线条的颜色和风格
plt.plot（）函数接受可用于指定这些参数的其他参数。要调整颜色，可以使用color关键字，它接受一个字符串参数，该参数实际上表示任何可想象的颜色。颜色可以通过多种方式指定：

```python
plt.plot(x, np.sin(x - 0), color='blue')        # specify color by name
plt.plot(x, np.sin(x - 1), color='g')           # short color code (rgbcmyk)
plt.plot(x, np.sin(x - 2), color='0.75')        # Grayscale between 0 and 1
plt.plot(x, np.sin(x - 3), color='#FFDD44')     # Hex code (RRGGBB from 00 to FF)
plt.plot(x, np.sin(x - 4), color=(1.0,0.2,0.3)) # RGB tuple, values 0 to 1
plt.plot(x, np.sin(x - 5), color='chartreuse'); # all HTML color names supported
```

线条风格用linestyle指定

```python
plt.plot(x, x + 0, linestyle='solid')
plt.plot(x, x + 1, linestyle='dashed')
plt.plot(x, x + 2, linestyle='dashdot')
plt.plot(x, x + 3, linestyle='dotted');

# For short, you can use the following codes:
plt.plot(x, x + 4, linestyle='-')  # solid
plt.plot(x, x + 5, linestyle='--') # dashed
plt.plot(x, x + 6, linestyle='-.') # dashdot
plt.plot(x, x + 7, linestyle=':');  # dotted
```

```python
#将线条颜色和风格结合
plt.plot(x, x + 0, '-g')  # solid green
plt.plot(x, x + 1, '--c') # dashed cyan
plt.plot(x, x + 2, '-.k') # dashdot black
plt.plot(x, x + 3, ':r');  # dotted red
```

### 设置坐标轴范围

```python
plt.plot(x, np.sin(x))

plt.xlim(-1, 11)
plt.ylim(-1.5, 1.5);
```

```python
#如果出于某种原因，您希望其中一个轴显示刻度反过来，则可以简单地反转参数的顺序。
plt.plot(x, np.sin(x))

plt.xlim(10, 1)
plt.ylim(-1.5, 1.5);
```

```python
#plt.axis（）还能自动收紧当前界面
plt.plot(x, np.sin(x))
plt.axis('tight')
```

```python
#确保横纵轴相同的比例
plt.plot(x, np.sin(x))
plt.axis('equal')
```

# 生成图例

plt.legend()

```python
plt.plot(x, np.sin(x), '-g', label = 'sin(x)')
plt.plot(x, np.cos(x), ':b', label = 'cos(x)')
plt.axis('equal')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.title('Curve')
plt.legend()
```

### Matplotlib Gotchas

虽然大多数plt函数直接转换为ax方法（例如plt.plot（）→ax.plot（）、plt.legend（）→ax.legend（）等），但并非所有命令都是如此。特别是，设置限制、标签和标题的功能稍加修改。对于matlab风格的函数和面向对象的方法之间的转换，进行以下更改

```python
ax = plt.axes()
ax.plot(x, np.sin(x))
ax.set(xlim=(0, 10), ylim=(-2,2),
       xlabel = 'x', ylabel = 'y',
      title = 'plot')
```
